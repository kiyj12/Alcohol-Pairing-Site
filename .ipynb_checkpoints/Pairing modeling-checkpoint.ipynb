```python
import numpy as np
import pandas as pd

df = pd.read_excel("../data/beer_pairing.xlsx", header=1)

df
```

```python
df = df.drop([df.columns[0], df.columns[1]], axis=1)

df
```

```python
def pairing(alcohol, number = 20):
    df_a = df.loc[:, ['Unnamed: 2', alcohol]]
    df_a = df_a.sort_values(by=[alcohol], ascending=False).values
    df_r = pd.DataFrame(df_a[:number], columns = ['ingredient', alcohol])
    
    return df_r

pairing('Blonde Ale')
```
